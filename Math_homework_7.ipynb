Задача1 (3балла). Найдите все экстремумы функции $f(x,y,z)=x+2y+3z$ при условии $\ln x+\ln y+\ln z=0$ и классифицируйте их.

Определим функцию Лагранжа:

$\mathcal{L}(x,y,z,\lambda)=x+2y+3z-\lambda(\ln x + \ln y + \ln z)$

Найдём частные производные функции $\mathcal{L}$ по переменным $x,y,z$, и $\lambda$, и приравняем их к нулю.

$\frac{\partial\mathcal{L}}{\partial x}=1-\frac{\lambda}{x}=0\implies\lambda=x$

$\frac{\partial\mathcal{L}}{\partial y}=2-\frac{\lambda}{y}=0\implies\lambda=2y$

$\frac{\partial\mathcal{L}}{\partial z}=3-\frac{\lambda}{z}=0\implies\lambda=3z$

$\lambda=x=2y=3z$

$y=\frac{x}{2},\quad z=\frac{x}{3}$

$\frac{\partial\mathcal{L}}{\partial\lambda}=-(\ln x+\ln y+\ln z)=0$

$\ln x+\ln y+\ln z=0$

$\ln(xyz)=0\implies xyz=1$

$x\cdot\frac{x}{2}\cdot\frac{x}{3}=1=\frac{x^3}{6}\implies x^3=6$

$x=\sqrt[3]{6}$

$y=\frac{x}{2}=\frac{\sqrt[3]{6}}{2}$

$z=\frac{x}{3}=\frac{\sqrt[3]{6}}{3}$

Итоговые значения:

$x=\sqrt[3]{6},\quad y=\frac{\sqrt[3]{6}}{2},\quad z=\frac{\sqrt[3]{6}}{3}$

Рассчитаем вторые частные производные:

$\frac{\partial^2\mathcal{L}}{\partial x^2}=\frac{\partial}{\partial x}\left(1-\frac{\lambda}{x}\right)=\frac{\lambda}{x^2}=\frac{x}{x^2}=\frac{1}{x}$

$\frac{\partial^2\mathcal{L}}{\partial y^2}=\frac{\partial}{\partial y}\left(2-\frac{\lambda}{y}\right)=\frac{\lambda}{y^2}=\frac{2y}{y^2}=\frac{2}{y}$

$\frac{\partial^2\mathcal{L}}{\partial z^2}=\frac{\partial}{\partial z}\left(3-\frac{\lambda}{z}\right)=\frac{\lambda}{z^2}=\frac{3z}{z^2}=\frac{3}{z}$

Используя полученные значения вторых частных производных, построим матрицу Гессе для функции $f(x,y,z)$:

$H=\begin{bmatrix}\frac{\partial^2\mathcal{L}}{\partial x^2}&0&0\\0&\frac{\partial^2\mathcal{L}}{\partial y^2}&0\\0&0&\frac{\partial^2\mathcal{L}}{\partial z^2}\end{bmatrix}=\begin{bmatrix}\frac{1}{\sqrt[3]{6}}&0&0\\0&\frac{2}{\sqrt[3]{6^2}}&0\\0&0&\frac{3}{\sqrt[3]{6^3}}\end{bmatrix}$

$\frac{1}{\sqrt[3]{6}}>0,\quad\frac{2}{\sqrt[3]{6^2}}>0,\quad\frac{3}{\sqrt[3]{6^3}}>0$

Все собственные значения матрицы Гессе положительны, это означает, что функция имеет локальный минимум в критической точке.

Задача 3 (3 балла). Найдите минимум функции $f(x,y,z)=x^2+3y^2+5z^2$, при ограничении $x+y+z≤−23$.

Определим функцию Лагранжа:

$\mathcal{L}(x,y,z,\lambda)=x^2+3y^2+5z^2+\lambda(x+y+z+23)$
где $\lambda$ — множитель Лагранжа.

Найдём частные производные функции $\mathcal{L}$ и приравняем их к нулю:

$\frac{\partial\mathcal{L}}{\partial x}=2x+\lambda=0\implies\lambda=-2x$

$\frac{\partial\mathcal{L}}{\partial y}=6y+\lambda=0\implies\lambda=-6y$

$\frac{\partial\mathcal{L}}{\partial z}=10z+\lambda=0\implies\lambda=-10z$

$\lambda=-2x=-6y=-10z$

$-2x=-6y\implies x=3y$

$-6y=-10z\implies y=\frac{5}{3}z$

Подставим $x=3y$ в ограничение $x+y+z=-23$:

$3y+y+z=-23$

$4y+z=-23$

Подставим $y=\frac{5}{3}z$:

$4\left(\frac{5}{3}z\right)+z=-23$

$\frac{20}{3}z+z=-23$

$\frac{20z+3z}{3}=-23$

$23z=-69\implies z=-3$

Найдём $y$ и $x$:

$y=\frac{5}{3}z=\frac{5}{3}(-3)=-5$

$x=3y=3(-5)=-15$

Вычислим минимальное значение функции:

$f(x,y,z)=x^2+3y^2+5z^2$

$f(-15,-5,-3)=(-15)^2+3(-5)^2+5(-3)^2=225+3\cdot25+5\cdot9=225+75+45=345$
