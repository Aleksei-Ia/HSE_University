Задача 2 (2 балла). У любой бомбы на диком западе есть шнур длиной 1 метр. Ковбой Боб случайным образом делает на шнуре два разреза. С какой вероятностью хотя бы один из получившихся кусков будет длиннее 0.5 метра?

------------------------------------------

Обозначим точки разрезов как $x$ и $y$, где $0\leq x\leq y\leq 1$. После разрезов длины кусков составляют:

Первый кусок: $x$

Второй кусок: $y-x$

Третий кусок: $1-y$

Хотя бы один из трех кусков должен быть больше $0.5$:

$x>0.5\quad или\quad y-x>0.5 \quad или\quad 1-y>0.5$

Разрезы можно интерпретировать как точки $(x,y)$ на плоскости $[0,1] \times [0,1]$ с условием $x\leq y$. Эта область представляет собой половину квадрата, треугольник с вершинами $(0,0)$,$(1,1)$ и $(0,1)\implies$ Площадь треугольника $=\frac{1}{2}$

$x\leq0.5$,

$y-x\leq0.5$, что эквивалентно $y\leq x+0.5$,

$1-y\leq 0.5$, что эквивалентно $y\geq0.5$

Выполнение всех этих условий задает треугольник с вершинами $(0,0.5)$, $(0.5,0.5)$ и $(0.5,1)$:

Площадь этого треугольника = $\frac{1}{2}\times(0.5-0)\times(1-0.5)=\frac{1}{8}$

Вероятность того, что хотя бы один кусок будет длиннее 0.5 =

$1-\frac {Площадь треугольника, где все куски\leq 0.5}{Общая площадь треугольника}=1-\frac{\frac{1}{8}}{\frac{1}{2}}=1-\frac{1}{4}=\frac{3}{4}=75\%$

Задача 3 (5 балла). Вероятность застать шерифа в городе зависит от двух факторов: первый — намечается ли в городе стычка, и второй — приедут ли молоденькие девушки из соседнего поселения продавать свой урожай на местную ярмарку.

Данная вероятность равна 0.18, если девушки не приедут и стычка не намечается, 0.9 — если стычка намечается и девушки приедут. 0.54 — если намечается только стычка и 0.36 — если приедут девушки, но стычка не намечается. Стычки происходят с вероятностью 0.4, а девушки приезжают с вероятностью 0.6

Стычки в городе никак не зависят от девушек, несмотря на то, что у каждой из них есть по пушке и ножу.

------------------------------------------

1. Рассмотрим вероятность \( P(A) \), где событие \( A \) описывает, что в городе происходит что-то (либо стычка, либо девушки привезли урожай).

$P(A) = P(B) + P(C) - P(B \cap C)$

где:

$P(B) = 0.6$ — вероятность того, что девушки приехали

$P(C) = 0.4$ — вероятность того, что началась стычка

$P(B \cap C)$ — вероятность того, что девушки приехали и началась стычка

Так как события \( B \) и \( C \) независимы, то:

$P(B \cap C) = P(B) \cdot P(C) = 0.6 \cdot 0.4 = 0.24$

Подставляем значения:

$P(A) = 0.6 + 0.4 - 0.24 = 0.76$

Итоговая вероятность:

$P(A) = 0.76$

2. Рассмотрим вероятность \( P(B \cap C) \), где событие \( B \cap C \) означает, что девушки привезли урожай и одновременно с этим началась стычка.

$P(B \cap C) = P(B) \cdot P(C)$

где:
$P(B) = 0.6$ — вероятность того, что девушки приехали
$P(C) = 0.4$ — вероятность того, что началась стычка

Так как события \( B \) и \( C \) независимы:

$P(B \cap C) = 0.6 \cdot 0.4 = 0.24$

Итоговая вероятность:

$P(B \cap C) = 0.24$

$P(D) = P(\text{Шериф в городе} | \text{Девушки и стычка}) \cdot P(\text{Девушки и стычка})$

где:

4. Рассмотрим вероятность \( P(\text{Девушки} | \text{Шериф}) \), где событие \( \text{Девушки} \) означает, что девушки приехали в город на ярмарку, а событие \( \text{Шериф} \) — что шериф находится в городе.

Используем формулу Байеса:

$P(Девушки | \text{Шериф}) = \frac{P(\text{Шериф и девушки})}{P(\text{Шериф})$

Где:

$P(\text{Шериф и девушки}) = P(\text{Шериф} | \text{Девушки и стычка}) \cdot P(\text{Девушки и стычка}) + P(\text{Шериф} | \text{Девушки и нет стычки}) \cdot P(\text{Девушки и нет стычки})$

$P(\text{Шериф}) = P(\text{Шериф} | \text{Девушки и стычка}) \cdot P(\text{Девушки и стычка}) + P(\text{Шериф} | \text{Девушки и нет стычки}) \cdot P(\text{Девушки и нет стычки}) + \ldots$

Рассчитаем \( P(\text{Шериф и девушки}) \):

$P(\text{Девушки и стычка}) = P(\text{Девушки}) \cdot P(\text{Стычка}) = 0.6 \cdot 0.4 = 0.24$

$P(\text{Девушки и нет стычки}) = P(\text{Девушки}) - P(\text{Девушки и стычка}) = 0.6 - 0.24 = 0.36$

Условные вероятности:

$P(\text{Шериф} | \text{Девушки и стычка}) = 0.54, \quad P(\text{Шериф} | \text{Девушки и нет стычки}) = 0.36$

Подставляем:

$P(\text{Шериф и девушки}) = 0.54 \cdot 0.24 + 0.36 \cdot 0.36 = 0.1296 + 0.1296 = 0.2592$

Рассчитаем \( P(\text{Шериф}) \), включая все возможные случаи:

$P(\text{Шериф}) = P(\text{Шериф} | \text{Девушки и стычка}) \cdot P(\text{Девушки и стычка}) + P(\text{Шериф} | \text{Девушки и нет стычки}) \cdot P(\text{Девушки и нет стычки}) + P(\text{Шериф} | \text{Нет девушек и стычка}) \cdot P(\text{Нет девушек и стычка}) + P(\text{Шериф} | \text{Нет девушек и нет стычки}) \cdot P(\text{Нет девушек и нет стычки})$

После подстановки и вычисления (аналогично выше):

$P(\text{Шериф}) = 0.2592 + \ldots = \text{значение}$

Итог:

$P(\text{Девушки} | \text{Шериф}) = \frac{0.2592}{P(\text{Шериф})}$

$P(S)=P(S\cap A^c,B^c)P(A^c,B^c)+P(S\cap A,B^c)P(A,B^c)+P(S\cap A^c,B)P(A^c,B)+P(S\cap A,B)P(A,B)$

$P(S)=P(S\cap A^c,B^c)P(A^c,B^c)+P(S\cap A,B^c)P(A,B^c)+P(S\cap A^c,B)P(A^c,B)+P(S\cap A,B)P(A,B)$
