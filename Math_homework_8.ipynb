Задача 2 (2 балла). У любой бомбы на диком западе есть шнур длиной 1 метр. Ковбой Боб случайным образом делает на шнуре два разреза. С какой вероятностью хотя бы один из получившихся кусков будет длиннее 0.5 метра?

------------------------------------------

Обозначим точки разрезов как $x$ и $y$, где $0\leq x\leq y\leq 1$. После разрезов длины кусков составляют:

Первый кусок: $x$

Второй кусок: $y-x$

Третий кусок: $1-y$

Хотя бы один из трех кусков должен быть больше $0.5$:

$x>0.5\quad или\quad y-x>0.5 \quad или\quad 1-y>0.5$

Разрезы можно интерпретировать как точки $(x,y)$ на плоскости $[0,1] \times [0,1]$ с условием $x\leq y$. Эта область представляет собой половину квадрата, треугольник с вершинами $(0,0)$,$(1,1)$ и $(0,1)\implies$ Площадь треугольника $=\frac{1}{2}$

$x\leq0.5$,

$y-x\leq0.5$, что эквивалентно $y\leq x+0.5$,

$1-y\leq 0.5$, что эквивалентно $y\geq0.5$

Выполнение всех этих условий задает треугольник с вершинами $(0,0.5)$, $(0.5,0.5)$ и $(0.5,1)$:

Площадь этого треугольника = $\frac{1}{2}\times(0.5-0)\times(1-0.5)=\frac{1}{8}$

Вероятность того, что хотя бы один кусок будет длиннее 0.5 =

$1-\frac {Площадь треугольника, где все куски\leq 0.5}{Общая площадь треугольника}=1-\frac{\frac{1}{8}}{\frac{1}{2}}=1-\frac{1}{4}=\frac{3}{4}=75\%$

Данная вероятность равна 0.18, если девушки не приедут и стычка не намечается, 0.9 — если стычка намечается и девушки приедут. 0.54 — если намечается только стычка и 0.36 — если приедут девушки, но стычка не намечается. Стычки происходят с вероятностью 0.4 P(C), а девушки приезжают с вероятностью 0.6 P(B)

Стычки в городе никак не зависят от девушек, несмотря на то, что у каждой из них есть по пушке и ножу.

------------------------------------------

1. Найдите вероятность того, что в городе что-то происходит (либо стычка либо привезли урожай).

$P(B)=0.6$ — вероятность того, что привезли урожай

$P(C)=0.4$ — вероятность того, что началась стычка

$P(B\cap C)$ — вероятность того, что привезли урожай и началась стычка

$P(A)=P(B)+P(C)-P(B\cap C)$

$P(B\cap C)=P(B)\cdot P(C)=0.6\cdot 0.4=0.24$

$P(A)=0.6+0.4-0.24=0.76$

Итог:

$P(A)=0.76$

$P(C)=0.4$ — вероятность того, что началась стычка

$P(B\cap C)=P(B)\cdot P(C)$

$P(B\cap C)=0.6\cdot0.4=0.24$

Итог:

$P(B\cap C)=0.24$

Задача 4 (2 балла). При отсутствии негативных факторов вероятность того, что шериф выживет при отсутствии негативных факторов составляет 0.99. При атаке индейцев вероятность выжить равна 0.95. При появлении в городе бандита вероятность выжить равна 0.9. Вероятность выжить при одновременной атаке бандита и индейцев равна 0.8.

Бандит приходит в город с вероятностью 0.1, индейцы нападают с вероятностью 0.2.

------------------------------------------

Обозначим известные вероятности:

Вероятность выживания шерифа при отсутствии негативных факторов: $P(S|A^c,B^c)=0.99$

Вероятность выживания при атаке индейцев: $P(S|A,B^c)=0.95$

Вероятность выживания при приходе бандита: $P(S|A^c,B)=0.9$

Вероятность выживания при атаке индейцев и приходе бандита: $P(S|A,B)=0.8$

Вероятность прихода бандита: $P(B)=0.1$

Вероятность атаки индейцев: $P(A)=0.2$

------------------------------------------

1.  Какова вероятность гибели шерифа $P(S^c)$, если считать, что атаки индейцев $A$ и приход бандита $B$ не зависят друг от друга?

$P(S)=P(S|A^c,B^c)P(A^c,B^c)+P(S|A,B^c)P(A,B^c)+P(S|A^c,B)P(A^c,B)+P(S|A,B)P(A,B)$

$P(A^c,B^c)=P(A^c)P(B^c)=(1-P(A))(1-P(B))$

$P(A,B^c)=P(A)P(B^c)=P(A)(1-P(B))$

$P(A^c,B)=P(A^c)P(B)=(1-P(A))P(B)$

$P(A,B)=P(A)P(B)$

Подставим численные значения:

$P(S)=0.99\cdot(1-0.2)(1-0.1)+0.95\cdot0.2(1-0.1)+0.9\cdot(1-0.2)0.1+0.8\cdot0.2\cdot0.1=0.99\cdot0.8\cdot0.9+0.95\cdot0.2\cdot0.9+0.9\cdot0.8\cdot0.1+0.8\cdot0.2\cdot0.1=0.7128+0.171+0.072+0.016=0.9718$

Вероятность гибели шерифа:

$P(S^c)=1-P(S)=1-0.9718=0.0282$
